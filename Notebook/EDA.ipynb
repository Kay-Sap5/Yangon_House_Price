```python
yangon = final_df[final_df['State'] == 'Yangon']
```

```python
yangon.drop(columns = ['State'] , inplace = True)
```

```python
count = yangon['city'].value_counts()
```

```python
def filter_city(x):
    x = str(x.strip())
    if x in count[count>50].index:
        return x
    else:
        return 'Others'
```

```python
yangon['city'] = yangon['city'].apply(filter_city)
```
