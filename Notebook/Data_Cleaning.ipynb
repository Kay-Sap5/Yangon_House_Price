## Converting Key_Value Data

```python
df['key_value_temp'] = df['key_value'].str.replace("[","").str.replace("]","")
```

```python

data = []

for i in df['key_value_temp'].str.split("),",regex = False):
    key = []
    value =[]
    for j in i:
        k = j.split(",")[0].replace("('",'').replace("'",'').strip()
        val = j.split(",")[1].replace("'",'').replace('"','').strip()
        key.append(k)
        value.append(val)
    data.append(dict(zip(key,value)))
key_value_df = pd.DataFrame(data)
    
        
key_value_df_back = key_value_df.copy()  
```

# Cleaning Area Column

In this column we have three kind of value type:
- foot square
- acre
- x_y(X x Y)
- after extracting all data i convert it into '0' and deleted

```python
key_value_df['Area'].unique()
```

```python
key_value_df[key_value_df['Area'] == '1045 ft<sup>2</sup>']
```

```python
def get_ft(x):
    if "ft" in x:
        return(x)
    else:
        return '0'

def make_ft_zero(x):
        if "ft<sup>2</sup>" in x:
            return('0')
        else:
            return x
    
```

```python
key_value_df['Area_ft'] = key_value_df['Area'].apply(get_ft)
```

```python
key_value_df['Area_ft'] = key_value_df['Area_ft'].str.replace('ft<sup>2</sup>',"")
```

```python
key_value_df.head()
```

```python
key_value_df['Area'] = key_value_df['Area'].apply(make_ft_zero)
```

```python
key_value_df['x_y'] = key_value_df['Area'].apply(lambda x: x if "x"  in str(x) else '0').str.strip()
```

```python
key_value_df['acre']= key_value_df['Area'].apply(lambda x: x if 'acre'  in str(x) else '0').str.strip()
```

```python
key_value_df
```

```python
def make_x_y_and_acre_zero(x):
    if 'x'  in x:
        return "0"
    elif 'acre' in x:
        return '0'
    else:
        return x
```

```python
key_value_df['Area'] = key_value_df['Area'].apply(make_x_y_and_acre_zero)
```

```python
key_value_df['Area'].unique()
```

```python
key_value_df.head()
```

```python
key_value_df.drop(columns = ['Area'] , inplace = True)
```

```python
# Transforming Floor Long Text Into Real value , ''
def floor_str_to_int(x):
    key = list(floor_trans.keys())
    value = list(floor_trans.values())
    
    x = str(x).strip()
    if x ==str(np.nan) or x =="":
        return float(0)
        
    elif x in key:
        ind = key.index(x)
        floor = value[ind]
        return float(np.round(floor,1))
```

```python
key_value_df['Floor'] = key_value_df['Floor'].apply(floor_str_to_int)
```
